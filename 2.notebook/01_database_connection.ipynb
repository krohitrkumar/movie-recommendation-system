```python
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
import os
```

```python
load_dotenv()

engine = create_engine(
    f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@"
    f"{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
engine
```

## IMPORTING MOVIES TABLE

```python
movies_df = pd.read_sql("SELECT * FROM movies", engine)
```

## IMPORTING RATINGS TABLE

```python
ratings_df = pd.read_sql("SELECT * FROM ratings", engine)
```

```python
## IMPORTING TAGS TABLE
```

```python
tags_df = pd.read_sql("SELECT * FROM tags", engine)
```
